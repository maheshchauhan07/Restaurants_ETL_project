```python
def convert_to_usd(price):
    if not price.startswith('$'):
        try:
            value = float(price)
            usd = value / 82  # Replace with the appropriate conversion rate
            return '${:,.2f}'.format(usd)
        except ValueError:
            pass
    return price
```

```python
df['max_price']=df['max_price'].apply(convert_to_usd)
df['min_price']=df['min_price'].apply(convert_to_usd)
```

```python
df['min_price']=df['min_price'].apply(lambda x:x.strip('$€₹'))
df['max_price']=df['max_price'].apply(lambda x:x.replace('$','').replace('€','').replace('₹',''))
```

```python
df['max_price'] = pd.to_numeric(df['max_price'], errors='coerce')
df['min_price'] = pd.to_numeric(df['min_price'], errors='coerce')
```
