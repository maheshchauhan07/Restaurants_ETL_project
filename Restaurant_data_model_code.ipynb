```python
import pandas as pd 
df=pd.read_csv('usa_transformed.csv')
df=df.head(100000)
df1=pd.read_csv('mexico_transformed.csv')
df1=df1.head(100000)
df2=pd.read_csv('canada_transformed.csv')
df2=df2.head(100000)
```

```python
main=pd.concat([df,df1,df2])
```

```python
pd.set_option('display.max_columns',None)
# main.to_csv('restaurants_data.csv')
```

```python
main=main.drop(columns='Unnamed: 0')
main['Country']=main['Country'].replace('USA_1','USA')
main.drop(columns=['Source','Restaurant_url'],inplace=True)
main['Restaurant_id']=main['Restaurant_id'].apply(lambda x: x.replace('d',''))
```

```python
fact_table=main[['Restaurant_id','reviews','rating','food','service','value','atmosphere']].reset_index(drop=True)
```

```python
fact_table['id']=fact_table.index
```

```python
restaurant_dim=main[['name','address','City','Country','About','Latitude','Longitude']].reset_index(drop=True)
restaurant_dim['Restaurant_dim_id']=restaurant_dim.index
```

```python
restaurant_price_dim=main[['Pricing','price_range','min_price','max_price','Average_price']].reset_index(drop=True)
restaurant_price_dim['price_id']=restaurant_price_dim.index
```
